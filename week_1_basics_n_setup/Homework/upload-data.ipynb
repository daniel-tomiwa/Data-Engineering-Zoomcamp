```python
from time import time
```

```python
while True:
    try: 
        t_start = time()

        df = next(df_iter)

        df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
        df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
        
        df.to_sql(name='green_taxi_data', con=engine, if_exists='append')

        t_end = time()

        print('inserted another chunk, took %.3f second' % (t_end - t_start))
    
    except StopIteration:
                print("Finished ingesting data into the postgres database")
                break
```

```python
SELECT 
	CAST(g1.lpep_pickup_datetime AS DATE),
	COUNT(1)
FROM green_taxi_data as g1
WHERE CAST(g1.lpep_pickup_datetime AS DATE) = CAST(g1.lpep_dropoff_datetime AS DATE)
GROUP BY CAST(g1.lpep_pickup_datetime AS DATE);


SELECT 
	MAX(g1.trip_distance) AS trip_sum,
	CAST(g1.lpep_dropoff_datetime AS DATE)
FROM 
	green_taxi_data AS g1
WHERE 
	CAST(g1.lpep_dropoff_datetime AS DATE) = CAST(g1.lpep_pickup_datetime AS DATE)
GROUP BY 
	CAST(g1.lpep_dropoff_datetime AS DATE)
ORDER BY
	trip_sum DESC;


SELECT COUNT(*)
FROM green_taxi_data AS g1
WHERE 
	g1.passenger_count = 2
AND 
	CAST(g1.lpep_pickup_datetime AS DATE) = '2019-01-01';


SELECT 
	gtd."PULocationID", gtd."DOLocationID", gtd."tip_amount"
FROM 
	green_taxi_data AS gtd
LEFT JOIN zones AS z1 ON gtd."PULocationID" = z1."LocationID"
LEFT JOIN zones AS z2 ON gtd."DOLocationID" = z2."LocationID"
WHERE gtd."PULocationID" = "Astoria Zone";


SELECT 
	z1."Zone" AS PU_Zone, z2."Zone" AS DO_Zone, gtd."tip_amount"
FROM 
	green_taxi_data AS gtd
LEFT JOIN zones AS z1 ON gtd."PULocationID" = z1."LocationID"
LEFT JOIN zones AS z2 ON gtd."DOLocationID" = z2."LocationID"
WHERE
	z1."Zone" = 'Astoria';
```
